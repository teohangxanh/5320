### Import the processed dataset from zipped file

```python
import gzip
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

with gzip.open("Preprocessed data.gz") as f:
    df = pd.read_csv(f)

df['Country Name'] = df['Country Name'].str.strip()
df.head()
```

```python
# List of "countries" that will be dropped
to_be_dropped = ['East Asia & Pacific (excluding high income)', 'Europe & Central Asia (excluding high income)', 'IDA only',
                 'Latin America & Caribbean (excluding high income)',
                 'Least developed countries: UN classification', 'Low & middle income',
                 'Low income', 'Upper middle income', 'South Asia'
                 'Lower middle income', 'Middle East & North Africa (excluding high income)',
                 'Middle income', 'Sub-Saharan Africa (excluding high income)']
df = df[~df['Country Name'].isin(to_be_dropped)]
```

```python
# Transform the top_10 list to a proper shape
filtered_df = df.groupby(by='Country Name').agg('sum').transpose() / 1e9
```

```python
# https://towardsdatascience.com/creating-bar-chart-race-animation-with-python-cdb01144074e
import bar_chart_race as bcr

bcr.bar_chart_race(df=filtered_df,
                   n_bars=10,
                   sort='desc',
                   title='Debts of top 10 countries from 1970 in Billion dollars',
                   filename='top_10.mp4')
```

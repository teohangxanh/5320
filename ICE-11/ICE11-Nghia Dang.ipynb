#### 1.2 Create a line chart which contains multiple lines. Analysis the data based on your line chart. Submit a screenshot of the graph which contains the selection object and a screenshot of your code (commented properly).


```python
import calendar
import numpy as np
import pandas as pd

# Create a fake data
regions = ["North", "South", "East", "West"]
months = list(range(1, 13))
df = pd.DataFrame(
    np.random.randint(1000, 2000, size=(len(months), len(regions))), columns=regions
)
df["month"] = months
# Convert month from number to abbreviation
df["month"] = df["month"].apply(lambda x: calendar.month_abbr[x])
# Set month as index
df.set_index("month", drop=True, inplace=True)
df = df.transpose()
df.head()
```

```python
plt.rcParams["figure.figsize"] = (15, 12)
df.transpose().plot()
```

#### 1.3 Use small multiples to separate the above chart into multiple subplots. Submit a screenshot of the multiple plots and a screenshot of your code (commented properly). Describe differences/similarities between these subplots. Analysis the data based on the plots.

```python
# make multiple subplots.
# you can define the size of figure and dpi (dot per inch, defalt dpi=72)
my_dpi = 50
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 15), dpi=my_dpi)
fig.supxlabel("Months", fontsize=35, color="magenta")
fig.supylabel("Sales", fontsize=35, color="magenta")

# title for entire figure
fig.suptitle("Sales of four regions in the entire year", fontsize=35, color="Yellow")

# edit subplots
axes[0, 0].set_title("North", fontsize=35, color="green")
axes[0, 0].plot(df.iloc[0, :])
axes[0, 1].set_title("South", fontsize=35, color="green")
axes[0, 1].plot(df.iloc[1, :])
axes[1, 0].set_title("East", fontsize=35, color="green")
axes[1, 0].plot(df.iloc[2, :])
axes[1, 1].set_title("West", fontsize=35, color="green")
axes[1, 1].plot(df.iloc[3, :])
```
